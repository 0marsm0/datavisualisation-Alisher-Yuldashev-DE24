# Matplotlib annotations

```python
import pandas as pd
import duckdb

df = pd.read_csv("data/1900-census-immigrant-pop-metro.csv")

df.head()
```

```python
df.info()
```

```python
df["Total foreign born"].iloc[0]
```

```python
df = df.fillna("0")
df.head()
```

```python
columns_to_change_type = (
    "Total foreign born",
    "Sweden",
    "Norway",
    "Denmark",
    "Finland",
)

for column in columns_to_change_type:
    df[column] = df[column].str.replace(",", "").astype(int)
df.info()
```

```python
df_nordic = duckdb.query(
    """--sql
        WITH nordic AS (
        FROM df 
            SELECT 
                state, 
                SUM("Total foreign born") as "Total foreigners", 
                SUM("Sweden") as Swedes, 
                SUM("Norway") as Norwegians, 
                SUM("Denmark") as Danish, 
                SUM("Finland") as Finnish
            GROUP BY
                state
        )
        FROM nordic n
        SELECT 
            *,
            n.swedes + n.norwegians + n.danish + n.finnish AS Nordic
        ORDER BY 
            nordic
        DESC
             """
).df()

df_nordic.head(10)
```

```python
import matplotlib.pyplot as plt

top_3_nordic = df_nordic.drop(["Total foreigners", "Nordic"], axis=1).set_index("State").iloc[:3]
top_3_nordic
```

```python
def nordic_plot():
    ax = top_3_nordic.plot(kind = 'bar', figsize=(12,6))
    ax.set_xticklabels(top_3_nordic.index, rotation = 0)

    fig = ax.get_figure()

    return fig, ax

nordic_plot()
```

```python
fig, ax = nordic_plot()

ax.containers
```

```python
ax.containers[0] 
```

## Annotate

```python
fig, ax = nordic_plot()

arrowprops = dict(arrowstyle = "->", linewidth = 2, connectionstyle="arc3, rad=.3")
ax.annotate(text = "Swedes", xy = (0.81, 35000), arrowprops=arrowprops, xytext=(0.82, 42000))
```

```python
# TODO for reader:
# draw other annotations 
# add labels and title
# remove legend 
# save the figure 

fig, ax = nordic_plot()

swe_bars, nor_bars, dan_bars, fin_bars = ax.containers


def annotate_nordic(bar_artist, text, x_offset, y_offset):

    xy = (bar_artist.get_x(), bar_artist.get_height())

    rad = 0.3 if xy[0] < 0 else -0.3

    arrowprops = dict(arrowstyle="->", linewidth=2, connectionstyle=f"arc3, rad={rad}")
    ax.annotate(
        text=text,
        xy=(xy[0] + 0.06, xy[1]),
        arrowprops=arrowprops,
        xytext=(xy[0] + x_offset, xy[1] + y_offset),
    )

    ax.legend().remove()


annotate_nordic(swe_bars[1], "Swedes", -0.1, 6000)
annotate_nordic(nor_bars[0], "Norwegians", 0.1, 6000)
annotate_nordic(dan_bars[1], "Danish", 0.1, 6000)
annotate_nordic(fin_bars[2], "Finnish", -0.06, 6000)

```
