## mplstyle and DRY for storytelling in matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import duckdb

# in python script
# Path(__file__).parent / "styles/base.mplstyle"

plt.style.use("styles/base.mplstyle")

df = pd.read_csv("data/jobs_in_data_2024.csv")

df.head()
```

```python
df.shape
```

```python

df_job_categories = df["job_category"].value_counts()
df_job_categories
```

## visualizations

```python
from charts_utils import horizontal_bar_options, save_fig_from_ax
from constants import BLUE_1, GRAY_1

colors = [GRAY_1] * len(df_job_categories)
colors[:4] = [BLUE_1] * 4

ax = df_job_categories.plot(
    kind="barh",
    color=colors
)

ax = horizontal_bar_options(
    ax, 
    title = "Number jobs by job category for data jobs in 2024", 
    xlabel = "# of jobs".upper(), 
    ylabel = "job category".upper()
)

save_fig_from_ax(ax, "figures/number_data_jobs.png")
```

```python
df_salary = (
    duckdb.query(
        """
    SELECT 
        job_category, 
        CAST(AVG(salary_in_usd) AS INT) AS avg_salary_usd
    FROM 
        df
    GROUP BY 
        job_category
    ORDER BY avg_salary_usd DESC             
"""
    )
    .df()
    .set_index("job_category")
)

df_salary
```

```python
from charts_utils import thousands_formatter

colors = [GRAY_1] * len(df_salary)
colors[3] = BLUE_1

# IMPORTANT to use pd.Series for colors to work (not Dataframe)
ax = df_salary["avg_salary_usd"].plot(kind="barh", color=colors, width=0.6)

ax = horizontal_bar_options(
    ax,
    xlabel="yearly salary in usd".upper(),
    ylabel="job category".upper(),
    title=r"$\bf{Data\ engineering}$"
    + " is both an extremely interesting and also\none that tops the salary"
)

ax = thousands_formatter(ax, axis="x")

save_fig_from_ax(ax, "figures/data_jobs_salaries.png")
```
